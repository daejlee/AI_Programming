```python
import torch

def normalize(t):
  result = t
  t -= t.min()
  div = t.max()
  t /= div
  return result

fl = torch.FloatTensor([[-3.2, 1.3, 123.5], [0.2, -5.2, 16.9]])
print(fl)
print(normalize(fl))
```

```python
def thresholding(t, threshold, value):
  result = t
  s = t.shape
  for i in range(s[0]):
    for j in range(s[1]):
      if t[i][j] <= threshold:
        t[i][j] = value
  return result

fl = torch.FloatTensor([[-3.2, 1.3, 123.5], [0.2, -5.2, 16.9]])
print(thresholding(fl, -3, -100))
print(thresholding(fl, 1.3, 0))
```

```python
import numpy as np
import matplotlib.pyplot as plt

fs=100; dt=1/fs; N=500
t=np.arange(0,N)*dt
x1t=1.0*np.sin(2*np.pi*1*t)
x2t=1/3*np.sin(2*np.pi*3*t)
x3t=1/5*np.sin(2*np.pi*5*t)
x4t=1/7*np.sin(2*np.pi*7*t)
x5t=1/9*np.sin(2*np.pi*9*t)
xt=x1t+x2t+x3t+x4t+x5t

df=fs/N
f=np.arange(0,N)*df
Xf=np.fft.fft(xt)

plt.subplot(5,1,1); plt.plot(t,xt,"y"); plt.ylim(-1,1); plt.grid()
plt.ylabel("x(t)")
plt.title("A periodic rect signal x(t)=x1(t)+x2(t)+x3(t)")
plt.subplot(5,1,2); plt.plot(t,x1t,"b"); plt.ylim(-1,1); plt.grid()
plt.ylabel("x1(t)")
plt.subplot(5,1,3); plt.plot(t,x2t,"b"); plt.ylim(-1,1); plt.grid()
plt.ylabel("x2(t)")
plt.subplot(5,1,4); plt.plot(t,x3t,"b"); plt.ylim(-1,1); plt.grid()
plt.ylabel("x3(t)")
plt.subplot(5,1,5); plt.plot(f[0:int(N/2+1)], np.abs(Xf[0:int(N/2+1)]), "b");
plt.xlabel("frequency(Hz)"); plt.ylabel("Spectrum X(f)"); plt.grid()
```
